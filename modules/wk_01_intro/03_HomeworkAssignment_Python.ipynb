## Setup -- DO NOT CHANGE THIS PART (Do Run it)

```python
import os
from pathlib import Path

# Helper function to display directory tree (similar to R's dir_tree)
def dir_tree(path=".", prefix=""):
    """Display a directory tree structure"""
    path = Path(path)
    print(f"{prefix}{path.name}/" if prefix else f"{path.resolve()}/")
    
    items = sorted(path.iterdir(), key=lambda x: (x.is_file(), x.name))
    for i, item in enumerate(items):
        is_last = i == len(items) - 1
        connector = "└── " if is_last else "├── "
        
        if item.is_dir():
            print(f"{prefix}{connector}{item.name}/")
            extension = "    " if is_last else "│   "
            # Recursively show subdirectories
            for subitem in sorted(item.iterdir(), key=lambda x: (x.is_file(), x.name)):
                subconnector = "    └── " if subitem == list(item.iterdir())[-1] else "    ├── "
                suffix = "/" if subitem.is_dir() else ""
                print(f"{prefix}{extension}{subconnector[4:]}{subitem.name}{suffix}")
        else:
            print(f"{prefix}{connector}{item.name}")

print("Setup complete!")
```
